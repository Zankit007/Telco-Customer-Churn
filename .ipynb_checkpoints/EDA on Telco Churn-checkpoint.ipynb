# Customer Churn in Telco - EDA 

Churn can be defined as customer attrition or customers leaving a Service Provider(SP). 
As the cost of retaining an existing customer is far less than onboaring a new customer, hence it becomes very important for businesses to understand why customers would want to leave and join the competition

In this notebook we will try an analyse what contributes towards customer churn for a Telco. 

Dataset used here is from Kaggle - [Telco-Customer-Churn](https://www.kaggle.com/blastchar/telco-customer-churn)

```python
#Making a Pie chart to visualize the churn % w.r.t Gender

fig, ax1 = plt.subplots(figsize=(10,8))

width = 0.5
data_pie = df_train.Churn.value_counts().sort_index()
ax1.pie(data_pie, 
        labels=['Retained', 'Churned'],
        autopct ='%1.1f%%',pctdistance=0.8,
        startangle=90, 
        textprops={'color':'white'},
        wedgeprops = {'width':width, 'edgecolor':'w'})

data_pie2 = df_train.groupby(['Churn', 'gender']).size().reset_index()
ax1.pie(
    data_pie2.iloc[:,2], 
    labels=list(data_pie2.gender),
    autopct='%1.1f%%',
    startangle=90,
    textprops={'color':'white', 'fontweight':'bold'},
    wedgeprops = {'width':width, 'edgecolor':'w'},
    radius=1-width,
    rotatelabels=True,
)
```

* The pie chart clearly shows that in Retained and Churned categories the Male and Females have almost equal values.
*  Through this we can conclude that Gender and Churn are not related

### <b>Tenure

```python
#Checking the min and max values from column
print("Min value in Tenure is {} and Max value is {}".format(df_train.tenure.min(), df_train.tenure.max()))
```

```python
#dividing tenure into bins across the tenure range and assigning lables
bins = [0,12,24,36,48,60,72]
labels = ['1yr', '2yr', '3yr', '4yr', '5yr', '6yr']
df_train['tenure_new'] = pd.cut(df_train['tenure'], bins, labels=labels)
```

```python
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(12,6))

sns.barplot(x = 'tenure_new', y = df_train.Churn_values*100, data = df_train, ax = ax1, ci= None)
ax1.set_xlabel("Tenure")
ax1.set_ylabel("% Churn")
ax1.set_ylim([0,100])
ax1.set_title("Tenure and Churn")


sns.barplot(x = 'tenure_new', y = df_train.Churn_values*100, hue='Contract' ,data = df_train, ax = ax2, ci= None)
ax2.set_xlabel("Tenure")
ax2.set_ylabel("% Churn")
ax2.set_ylim([0,100])
ax2.set_title("Tenure and Churn across Contracts")
```

* The customers having a shorter tenure tend to churn the most
* The churn descreases as the tenure of customer inceases
* From the contracts point of view, the monthly contracted customers leave the most, followed by one year and two year contracts respectively.
* So, if the service provider is able to retain new customers beyond the first year atleast their chances of leaving the SP would decrease.

### <b> Services

```python
Services = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
            'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
n_cols = 2
n_rows = len(Services)
fig = plt.figure(figsize=(15,40))
#fig.suptitle('Distribution of Service Types and relation with Churn')
idx = 0

for serviceType in enumerate(Services):
    # Fetch data of Service Type
    data = df_train[serviceType[1]].value_counts(normalize=True).sort_index()

    # Now, plot the data
    i = 0
    for i in range(n_cols):
        idx+=1
        axes = fig.add_subplot(n_rows, n_cols, idx)

        # On column 1 - Plot the data distribution on bar plot
        if idx%2 != 0:
            axes.bar(data.index, data*100, color=my_colors)
        # On column 2 - Plot the percentage of churns on each service type
        else:
            sns.barplot(x=serviceType[1], y=df_train.Churn_values*100, data=df_train, ci=None, ax=axes)

        if idx == 1 : axes.set_title('Distribution of service category data')
        if idx == 2 : axes.set_title('% of churn in each service category')
            
        axes.set_xlabel(serviceType[1])
        axes.set_ylabel('')
        axes.set_ylim(0,100)

fig.tight_layout()
plt.show()
```


- Phone Service - Almost all (90%) customers have Phone Service. However, the churn rate is very low compared to the volume (25%). On the other hand, churn rate is almost same (25%) among non-phone service customers as well (though they form only 10% of the population). Thus, selling Phone Service is more beneficial for the CSP.
- Multiple Lines - Among Phone Service users, 45% has Multiple Lines while the rest 45% have single line. Churn rate is slightly higher among users with multiple lines, but not that significant (28%).
- Internet Service - Approx 20% customers do not use internet. Among customers using internet, approx 45% use Fiber Optic and 35% use DSL. However, churn among Fiber Optic users is pretty high as well (> 40%), while it is 20% among DSL users. Hence, CSP needs to look into the quality of its internet service.
- Online Security - 50% customers do not use online security service, and churn rate among such users is highest as well (40%).
- Online Backup, Device Protection and Tech Support - Same as above, most of the customers do not use this service and they tend to churn as well.
- Streaming TV and Streaming Movies Among users with internet services (which is mandatory for availing these services), almost half use these streaming services. The rate of churn is more or less same as well (35%) irrespective of usage of streaming services.
